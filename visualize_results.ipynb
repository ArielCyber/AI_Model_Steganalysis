# Supervised Experiments

```python
import pandas as pd

chosen_zoo = "svhn" # mnist, cifar10, stl10, svhn
msb = False

df = pd.read_csv(f'results_supervised{"_msb" if msb else ""}.csv')
df_only_chosen_zoo = df[df["zoo"]==chosen_zoo]
df_lsb_top_clf = df_only_chosen_zoo.loc[df_only_chosen_zoo.groupby('lsb')['f1'].idxmax()]
df_lsb_top_clf.sort_values(by=['lsb'], inplace=False, ascending=[False])
#df.sort_values(by=['lsb', 'clf'], inplace=True, ascending=[False, True])
#df
#df[]
```
